```python
import pystan
import numpy as np
import pandas as pd

from scipy.integrate import ode
backend = 'dopri5'

# Plotting
%matplotlib inline
# Make inline plots raster graphics
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('pdf', 'svg')
import matplotlib.pyplot as plt
```

```python
# X = (X[0],X[1])
ode_rhs = lambda t, X, k: [X[1],-X[0]-k*X[1]] if t>5 else [X[1],-10*X[0]-k*X[1]]
```

```python
τs = np.arange(1,11,.1)
solver = ode(ode_rhs).set_integrator(backend)
solver.set_initial_value([0,1]).set_f_params(.15)
ys = [solver.integrate(τ)+np.random.normal(0,.05,2) for τ in τs]
plt.plot(τs,ys)
```

```python
sho_code = """
functions {
    real[] sho(real t, real[] y, real[] theta, real[] x_r, int[] x_i) {
        real dydt[2];
        real kappa;
        dydt[1] = y[2];
        if (t > 5)
            kappa = 1.0;
        else 
            kappa = 10.0;
        dydt[2] = -kappa*y[1] - theta[1] * y[2];
        return dydt; }
}
data {
    int<lower=1> T;
    real y[T,2];
    real t0;
    real ts[T];
}
transformed data {
    real x_r[0];
    int x_i[0];
}
parameters {
    real y0[2];
    vector<lower=0>[2] sigma;
    real theta[1];
}
model {
    real z[T,2];
    sigma ~ cauchy(0,2.5);
    theta ~ normal(0,1);
    y0 ~ normal(0,1);
    z = integrate_ode_rk45(sho, y0, t0, ts, theta, x_r, x_i);
    for (t in 1:T)
        y[t] ~ normal(z[t], sigma);
}
"""

sho_dat = {'T': len(ys),
               'y': ys,
               't0': 0,
               'ts': τs}

sm = pystan.StanModel(model_code=sho_code)
```

```python
fit = sm.sampling(data=sho_dat, iter=2000, chains=4)
```
